# Matrix conditions

```python
import numpy as np
```

```python
# Well-conditioned system
A1 = np.array([[1, 2],
               [1, 1]])
b1 = np.array([3, 2])

# Solve the system
x1 = np.linalg.solve(A1, b1)
print("Well-conditioned system solution (exact):", x1)

# Slight perturbation in b
b1_perturbed = np.array([3.00001, 2])
x1_perturbed = np.linalg.solve(A1, b1_perturbed)
print("Well-conditioned system solution (perturbed):", x1_perturbed)

# Ill-conditioned system
A2 = np.array([[1.00001, 1],
               [1,       1]])
b2 = np.array([2.00001, 2])

# Solve the system
x2 = np.linalg.solve(A2, b2)
print("Ill-conditioned system solution (exact):", x2)

# Slight perturbation in b
b2_perturbed = np.array([2, 2])
x2_perturbed = np.linalg.solve(A2, b2_perturbed)
print("Ill-conditioned system solution (perturbed):", x2_perturbed)
```

```python
import numpy as np

cond_A1 = np.linalg.cond(A1)
cond_A2 = np.linalg.cond(A2)

print("Condition number of A1 (well-conditioned):", cond_A1)
print("Condition number of A2 (ill-conditioned):", cond_A2)
```

Let's observe a smaller, more observable case:

```python
import numpy as np

# Set print precision
np.set_printoptions(precision=8, suppress=True)

# Symmetric matrix with an almost-zero row/column
A = np.array([
    [4, 2, 2, 0],
    [2, 4, 2, 0],
    [2, 2, 4, 0],
    [0, 0, 0, 0.0000001]
], dtype=float)

# Known solution
x_true = np.array([1, 1, 1, 0.0000001], dtype=float)

# Compute right-hand side
b = A @ x_true

# Print results
print("A =\n", A)
print("x_true = ", x_true)
print("b = ", b)
```

```python
# Solve the system
x = np.linalg.solve(A, b)
print("Original system solution:", x)

# Perturb the last b value slightly
b_perturbed = b.copy()
b_perturbed[3] = 0  # Remove tiny signal
x_perturbed = np.linalg.solve(A, b_perturbed)
print("Perturbed b solution:", x_perturbed)

# Remove last row/col to make 3x3 well-conditioned system
A_reduced = A[:3, :3]
b_reduced = b[:3]
x_reduced = np.linalg.solve(A_reduced, b_reduced)
print("Reduced 3x3 system solution:", x_reduced)

# Condition number before (4x4 system)
cond_full = np.linalg.cond(A, np.inf)
print("Condition number (4x4):", cond_full)

# Condition number after removing problematic row/col (3x3 system)
cond_reduced = np.linalg.cond(A_reduced, np.inf)
print("Condition number (3x3):", cond_reduced)
```

Random 3x3 example

```python
import numpy as np

# Ill-conditioned 3x3 matrix
A = np.array([
    [1, 1, 1],
    [1, 1.0001, 1],
    [1, 1, 1.0001]
], dtype=float)

# Original right-hand side
b = np.array([3, 3.0001, 3.0001])

# Solve original system
x = np.linalg.solve(A, b)

# Slight perturbation in b
b_perturbed = np.array([3, 3, 3.0001])

# Solve perturbed system
x_perturbed = np.linalg.solve(A, b_perturbed)

print("Original solution:", x)
print("Perturbed solution:", x_perturbed)
print("Difference:", x_perturbed - x)
```

```python
import numpy as np

# Creative ill-conditioned 3x3 matrix (symmetric, some zeros)
A = np.array([
    [10, 9.9999, 0],
    [9.9999, 10, 0.0001],
    [0, 0.0001, 0.000001]
], dtype=float)

# Known solution
x_true = np.array([1, 1, 1], dtype=float)

# Compute right-hand side
b = A @ x_true

# Solve original system
x = np.linalg.solve(A, b)

# Slight perturbation in b (tiny change in last entry)
b_perturbed = np.array([b[0], b[1], b[2] + 1e-7])

# Solve perturbed system
x_perturbed = np.linalg.solve(A, b_perturbed)

print("Original solution:", x)
print("Perturbed solution:", x_perturbed)
print("Difference:", x_perturbed - x)

# Condition number with infinity norm
print("Condition number (∞-norm):", np.linalg.cond(A, np.inf))
```

```python
import numpy as np

# Matrix as before
A = np.array([
    [10, 9.9999, 0],
    [9.9999, 10, 0.0001],
    [0, 0.0001, 0.000001]
], dtype=float)

# Known solution
x_true = np.array([1, 1, 1], dtype=float)
b = A @ x_true

# Compute SVD of A
U, s, Vt = np.linalg.svd(A)

# Smallest singular value and corresponding singular vector in U
u_min = U[:, -1]

# Perturb b *along* the most sensitive direction (u_min)
epsilon = 1e-5
b_perturbed = b + epsilon * u_min

# Solve original and perturbed systems
x = np.linalg.solve(A, b)
x_perturbed = np.linalg.solve(A, b_perturbed)

print("Original solution:", x)
print("Perturbed solution:", x_perturbed)
print("Difference:", x_perturbed - x)
print("Condition number (∞-norm):", np.linalg.cond(A, np.inf))
print("Smallest singular value:", s[-1])
```
